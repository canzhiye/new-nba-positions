```python
import psycopg2
import pandas as pd
import numpy as np
from sklearn.decomposition import TruncatedSVD
from sklearn import (manifold, datasets, decomposition, ensemble, lda, random_projection)


conn = psycopg2.connect("dbname=nbadata user=postgres")
cur = conn.cursor()
```

```python
cur.execute("SELECT * FROM shooting;")
data = cur.fetchall()
shooting = [s[10:] for s in data]
players_index = [p[4] for p in data]
# players_index
```

```python
df = pd.DataFrame(shooting)
df.columns = [c.name for c in cur.description][10:]
df.index = players_index
df = df.groupby(df.index).sum()
df
```

```python
df = df.convert_objects(convert_numeric=True)
df = df.astype(float)
corr = df.T.corr()
X = corr.as_matrix()
df.dtypes
```

```python
# perform t-SNE dimensionality reduction and print to file
X_reduced = TruncatedSVD(n_components=10, random_state=0).fit_transform(X)
tsne = manifold.TSNE(n_components=3, perplexity=40, verbose=2)
X_tsne = tsne.fit_transform(X_reduced)

df2 = pd.DataFrame(X_tsne)
df2['player'] = df.index
# df2['team'] = [x[1] for x in df.index]
df2.columns = ['x', 'y', 'z', 'player']
df2
```
